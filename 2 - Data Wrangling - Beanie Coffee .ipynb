## Customers

```python
customers.head()
```

```python
customers['age'] = dt.datetime.now().year - customers.birth_year
```

```python
customers
```

```python
customers.describe()
```

```python
customers = customers.drop(columns=['customer_first-name', 'customer_email',\
                                    'loyalty_card_number', 'birth_year'])
```

```python
purchases = sales.merge(customers, how='left', on='customer_id')
```

```python
purchases.head()
```

```python
purchases.isnull().sum()
```

```python
purchases.home_store.fillna(1, inplace=True)
purchases.gender.fillna('N', inplace=True)
purchases.age.fillna(purchases.age.mean(),inplace=True)
purchases.birthdate.fillna('2000-01-01', inplace=True)
purchases.customer_since.fillna(purchases.transaction_date, inplace=True)
purchases.customer_since = pd.to_datetime(purchases.customer_since)
```

```python
dt2019 = dt.date(2019,1,1)
customer_years = dt2019.year - purchases.customer_since.dt.year
purchases.customer_since = customer_years
```

```python
purchases['birth_month'] = pd.to_datetime(purchases.birthdate).dt.month
purchases['birth_day'] = pd.to_datetime(purchases.birthdate).dt.day
purchases['birthday'] = purchases['birth_month'].astype(str) + "-" + purchases['birth_day'].astype(str)
purchases = purchases.drop(columns=['birthdate', 'birth_month', 'birth_day'])
purchases
```

I have combined the customer data with the sales receipts data. Over half of the purchases made in April were made by customers who did not belong to the customer loyalty program. For the purposed of filling in null data, I've set any customer who was not a loyalty member to 0 years in the loyalty program, the average age of customers, unknown gender, and a January 1st Birthday.

In the next step, I will add the product types to the list of orders.

```python
products.head()
```

```python
products2 = products.drop(columns = ['product_description', 'unit_of_measure','current_wholesale_price', 'current_retail_price', 'tax_exempt_yn', 'promo_yn'])
purchases = purchases.merge(products2, how='left', on='product_id')
```

```python
purchases['transaction_date'] = pd.to_datetime(purchases['transaction_date'],format='%Y-%d-%b')
```

```python
purchases['day_of_week'] = purchases['transaction_date'].dt.day_name()
```
